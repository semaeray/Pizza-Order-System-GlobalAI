# Yeni Bölüm

```python
#bir pizza üst sınıfı oluşturuyoruz ve menüde yazan pizzaları alt sınflar olarak belirliyoruz.
class Pizza:
    def __init__(self,name, contents, price):     
        self.contents = contents                   
        self.price = price

    def get_description(self):
        return self.contents
        
    def get_cost(self):
        return self.price

          
#Margarita,Klasik,Türk ve Sade olmak üzere menüde yer alan pizza alt sınıflarını oluşturduk.

class Margarita(Pizza):
    def __init__(self):
        super().__init__("Margarita Pizza","Keçi Peyniri ve Mantar",45)



class Classic(Pizza):
    def __init__(self):
        super().__init__("Klasik Pizza","Zeytin,Soğan ve Keçi Peyniri",50)



class Plain(Pizza):
    def __init__(self):
        super().__init__("Sade Pizza","peynir ve mısır",60)



class Turk(Pizza):
    def __init__(self):
        super().__init__("Türk Pizza","et,soğan ve peynir",70)


klasik=Classic()
plain=Plain()
turk=Turk()
margarita=Margarita()


#Kullanıcının pizza seçmesi için bir fonksiyon tanımladık ve hesapla1 listesine kullanıcının seçtiği pizzanın fiyatını eklerken, tercihler listesine seçtiği pizza numarasını
#ekledik. Böylece kullanıcı siparişi tamamlarken hangi pizzayı seçtiğini görebilecek.

hesapla1=[]
def pizzatercih():                               
  tercihler=[]                         
  global hesapla1
  print("Pizza seçenekleri: \n 1.Klasik pizza \n 2. Sade Pizza \n 3. Türk Pizzası \n 4.Margarita Pizza")
  while True:
        tercih = input("Hangi pizzayı istersiniz? (1-4)")
        

        if tercih == "1":
            pizza = klasik
            print("Klasik pizza servis ediliyor")
            tercihler.append("1")
            hesapla1.append(klasik.get_cost())
            print("tercih ettiğiniz pizza:")
            print(tercihler)
            break
        elif tercih == "2":
            pizza= plain
            print("Sade pizza servis ediliyor")
            tercihler.append("2")
            hesapla1.append(plain.get_cost())
            print("tercih ettiğiniz pizza:")
            print(tercihler)
            break
        elif tercih == "3":
            pizza = turk
            print("Türk Pizzası servis ediliyor")
            tercihler.append("3")
            hesapla1.append(turk.get_cost())
            print("tercih ettiğiniz pizza:")
            print(tercihler)
            break
        elif tercih == "4":
            pizza= margarita
            print("Margarita pizzası servis ediliyor")
            tercihler.append("4")
            hesapla1.append(margarita.get_cost())
            print("tercih ettiğiniz pizza:")
            print(tercihler)
            break
        
        else:
            print("Geçersiz seçim.")
            continue



#Pizza boyutunun seçilebilmesi için pizzasize adında bir üst sınıf oluşturuldu

class pizzasize():
   def __init__(self, contents, price):     
        self.contents = contents
        self.price = price
        
   def get_description(self):
        return self.contents
        
   def get_cost(self):
        return self.price

# Pizza boyutlarının alt sınıfları oluşturuldu.

class Buyukboy(pizzasize):
  def __init__(self):
        super().__init__("büyük boy",20)


class Ortaboy(pizzasize):
  def __init__(self):
        super().__init__("orta boy",15)


class Kucukboy(pizzasize):
  def __init__(self):
        super().__init__("kucuk boy",10)

kucukboy=Kucukboy()
ortaboy= Ortaboy()
buyukboy= Buyukboy()


#  Kullanıcının pizzasını istediği boyutta seçebilmesi için bir boyuttercih fonksiyonu oluşturduk.
#  hesapla2 listesine kullanıcının seçtiği pizza boyutunun fiyatını eklerken, tercihler listesine seçtiği pizza boyutunun numarasını 
#  ekledik. Böylece kullanıcı siparişi tamamlarken hangi pizza boyutunu seçtiğini görebilecek.

hesapla2=[]
def boyuttercih():
  tercihler=[]
  global hesapla2
  print("Boyut seçenekleri: \n 1.Küçük boy \n 2. Orta boy \n 3. Büyük boy")
  while True:
        tercih = input("Hangi boyut pizza istersiniz? (1-2-3)")
        

        if tercih == "1":
            boyut = kucukboy
            print("Küçük boy servis ediliyor")
            tercihler.append("1")
            hesapla2.append(kucukboy.get_cost())
            print("tercih ettiğiniz boyut:")
            print(tercihler)
            break
        elif tercih == "2":
            boyut = ortaboy
            print("Orta boy servis ediliyor")
            tercihler.append("2")
            hesapla2.append(ortaboy.get_cost())
            print("tercih ettiğiniz boyut:")
            print(tercihler)
            break
        elif tercih == "3":
            boyut= buyukboy
            print("Büyük boy servis ediliyor")
            tercihler.append("3")
            hesapla2.append(buyukboy.get_cost())
            print("tercih ettiğiniz boyut:")
            print(tercihler)
            break
  
        
        else:
            print("Geçersiz seçim.")
            continue


# Soslar için üst sınıf oluşturuldu.


class Soslar:
    def __init__(self, contents, price):     
        self.contents = contents
        self.price = price

    def get_description(self):
        return self.contents
        
    def get_cost(self):
        return self.price

#Menüde yazan soslara özel alt sınıflar oluşturuldu.

class Zeytin(Soslar):
    def __init__(self):
        super().__init__("Zeytin",5)
        self.name = "zeytin sosu"


class Mantar(Soslar):
    def __init__(self):
        super().__init__("Mantar",6)
        self.name="mantar sosu"



class Peynir(Soslar):
    def __init__(self):
        super().__init__("Peynir",7)
        self.name= "peynir sosu"



class Et(Soslar):
    def __init__(self):
        super().__init__("Et",10)
        self.name="et sosu"


class Sogan(Soslar):
    def __init__(self):
        super().__init__("Soğan",6)
        self.name="soğan sosu"


class Misir(Soslar):
    def __init__(self):
        super().__init__("Mısır",7)
        self.name = "mısır sosu"


zeytin=Zeytin()
mantar=Mantar()
peynir=Peynir()
et=Et()
sogan=Sogan()
misir=Misir()

#  Kullanıcının istediği sosu seçebilmesi için bir sostercih fonksiyonu oluşturduk.
#  hesapla3 listesine kullanıcının seçtiği sosun fiyatını eklerken, tercihler listesine seçtiği sosun numarasını 
#  ekledik. Böylece kullanıcı siparişi tamamlarken hangi sosu seçtiğini görebilecek.


hesapla3=[]
def sostercih():
  global hesapla3
  tercihler=[]
  print("Sos seçenekleri: \n 1.Zeytin sosu \n 2.Mantar sosu \n 3.Peynir sosus \n 4.Et sosu \n 5.Soğan sosu \n 6.Mısır sosu ")
  while True:
        tercih = input("Hangi sosu istersiniz? (1-6)")
        

        if tercih == "1":
            sos = zeytin
            print("zeytin sosu servis ediliyor")
            tercihler.append("1")
            hesapla3.append(zeytin.get_cost())
        elif tercih == "2":
            sos = mantar
            print("mantar sosu servis ediliyor")
            tercihler.append("2")
            hesapla3.append(mantar.get_cost())
        elif tercih == "3":
            sos = peynir
            print("peynir sosu servis ediliyor")
            tercihler.append("3")
            hesapla3.append(peynir.get_cost())
        elif tercih == "4":
            sos = et
            print("et sosu servis ediliyor")
            tercihler.append("4")
            hesapla3.append(et.get_cost())
        elif tercih == "5":
            sos = sogan
            print("soğan sosu servis ediliyor")
            tercihler.append("5")
            hesapla3.append(sogan.get_cost())
        elif tercih == "6":
            sos = misir
            print("mısır sosu servis ediliyor")
            tercihler.append("6")
            hesapla3.append(misir.get_cost())
        else:
            print("Geçersiz seçim.")
            continue

        devam = input("Başka sos ister misiniz? (E/H): ")
        if devam.lower() != "e":
          print("tercih ettiğiniz soslar:")
          print(tercihler)
          break


# Kullanıcının seçtiği pizzanın, pizza boyutunun ve sosların fiyatının hesaplandığı bir hesap fonksiyonu oluşturduk.

def hesap():
  hesapla=hesapla1+hesapla2+hesapla3
  hesap= sum(hesapla)
  print(hesap)

#Kullanıcının pizza siparişini başlatan ve seçim işlemlerini gerçekleştirdiği bir secim fonksiyonu oluşturduk.

def secim():
  pizzatercih()
  sostercih()
  boyuttercih()
  print("ödemeniz gereken hesap miktarı")
  hesap()
  


  
secim()

    
```

```python
# Kullanıcının siparişi onayladığı ve ödeme işlemlerini yaptığı bölümü hazırladık. Bu bölümde sipariş onayı verildiğinde kullanıcıdan kart bilgileri istenerek
# ödeme işlemi gerçekleştiriliyor.

siparis_onayi = input('Siparişi Onayla (y/n)')   #(y=yes,n=no)

if siparis_onayi == 'y':

  name = input('\nLütfen isminizi giriniz: ')
  while True:
        try:
            id_num = int(input('Lütfen TC kimlik numaranızı giriniz: '))
            if len(str(id_num)) == 11:
                break
            else:
                print('TC kimlik numarası 11 haneli ve rakamlardan oluşmalıdır\n')
        except:
            print('TC kimlik numarası 11 haneli ve rakamlardan oluşmalıdır\n')
  while True:
        try:
            credit_cart_num = int(input('Lütfen kredi kartı numaranızı giriniz: '))
            if len(str(credit_cart_num)) == 16:
                password=input("Lütfen kredi kartı şifrenizi giriniz:")
                break
            else:
                print('Kart numarası 16 haneli ve rakamlardan oluşmalıdır\n')

        except:
          print("Kart numarası 16 haneli ve rakamlardan oluşmaktadır.")
else:
    print('Seçiminiz iptal edildi')
    exit
```

```python
#  Kullanıcının kişisel ve sipariş bilgilerinin bulunduğu Orders_Database.csv dosyasını oluşturduk.

zaman = datetime.datetime.now().strftime("%Y-%m-%d %H:%M:%S")
hesap=sum(hesapla1+hesapla2+hesapla3)
yeni_siparis = [zaman, name, id_num, credit_cart_num, password,hesap]

with open('Orders_Database.csv', 'a') as dosya:
        db_writer = csv.writer(dosya)
        db_writer.writerow(['Tarih/Saat', 'İsim', 'T.C', 'Kredi Kartı No', 'Şifre','Tutar'])
        db_writer.writerow(yeni_siparis)
```
